```python
#filling blank name fields
working_dict = {}
for x in df.index:
    if(not pd.isna(df.loc[x, 'Name'])):
        working_dict[df.loc[x, 'Customer_ID']] = df.loc[x, 'Name']
print(working_dict)
for x in df.index:
    if(pd.isna(df.loc[x, 'Name'])): #if name is blank
        df.loc[x, 'Name'] = working_dict[df.loc[x, 'Customer_ID']]
```

```python
df['Age'] = df['Age'].apply(lambda x: remove_trailing_character(x, '_'))
df['Age'] = df['Age'].apply(lambda x: remove_leading_character(x, '-'))
```

```python
for x in df.index:
    working_dict[df.loc[x, 'Customer_ID']] = set()

for x in df.index:
    if(not pd.isna(df.loc[x, 'Credit_History_Age'])):
        years = ""
        months = ""
        space = 0
        for i in df.loc[x, 'Credit_History_Age']:
            #print(i)
            if(i.isnumeric()):
                if(space == 0):
                    years += i
                else:
                    months += i
            if(i == " "):
                space += 1
        space = 0
        working_dict[df.loc[x, 'Customer_ID']].add((int(years), int(months)))
        #print((years, months))
for x in working_dict:
    minimum = min(working_dict[x])
    maximum = max(working_dict[x])
    working_dict[x] = [minimum, maximum]
print(working_dict) #dict with 2 tuples in form (years, months) (one for minimum and one for maximum) of credit history age for every customer id
```

```python
for x in working_dict:
    working_dict[x] = 12*(working_dict[x][1][0]-working_dict[x][0][0]) + working_dict[x][1][1]- working_dict[x][0][1] #values are now number of months beetwen first and last credit
print(working_dict)
print(max(working_dict.values())) 
# =11, which shows that records for every customer are in span of 1 year, so in case of missing or wrong age value I can replace it with age value from different record
```
